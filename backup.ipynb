```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_data.csv")
# to know the first five values
print(df.head())
```

```python
# to know the description
df.describe()
```

```python
#TO know the information
df.info()
```

```python
#to find the null values
df.isnull().sum()
```

```python
# deleting coloumn
#drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
```

```python
print(df.head())
```

```python
# changing weekly studyhours data column by using string change function

df["WklyStudyHours"] = df["WklyStudyHours"].str.replace('5 - 10', "05 - 10")
df.head()
```

```python
# creating charts using seaborn

#gender distribution
plt.figure(figsize = (5,5))
gd = sns.countplot(data = df , x = "Gender" ,palette = "rainbow")
plt.title(" Gender Distribution")
plt.show
sns.set_style("whitegrid")
```

# from the above chat number of female are more than number of male.

```python
# finding is there any impact of parents eduction on students

#also creating groups by using .groupby function
```

```python
gb = df.groupby('ParentEduc').agg({"MathScore" : "mean", "ReadingScore" : "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Impact of parents eduction on students")
plt.show
sns.set_style("whitegrid")
```

# from the above chat(heatmap) we conclude that education of parents showing impact on Students

```python
# finding is there any impact of parents ParentMaritalStatus on students

ms = df.groupby('ParentMaritalStatus').agg({"MathScore" : "mean", "ReadingScore" : "mean", "WritingScore": "mean"})
print(ms)
```

```python
plt.figure(figsize = (5,5))
sns.set_style("whitegrid")
sns.heatmap(ms, annot = True)
plt.title("Impact of parents ParentMaritalStatus on students" )
plt.show
sns.set_style("whitegrid") #to changing style
```

# from the above chat(heatmap) we conclude that ParentMaritalStatus showing negligible impact on Students

```python
#Distribution of ethnic groups
```

```python
#to know the count by using pandas .loc function

groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ['groupA', 'groupB', 'groupC', 'groupD', 'groupE']
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = '%1.2f%%') #we used autopct for getting percentage " autopct = '%1.2f%%' "
plt.title("Distribution of Ethnic Group")
sns.set_style("whitegrid")
```

```python
ax = sns.countplot(data = df , x = "EthnicGroup" , palette = "rainbow")
ax.bar_label(ax.containers[0])
plt.title("Ethnic Group")
sns.set_style("whitegrid")
```
